```python
import random
from IPython.display import clear_output as co
session="y"
game='y'
fruits=['apple', 'melon', 'watermelon', 'banana', 'kiwi', 'orange', 'grape', 'pear', 'lemon',]
cars=['honda', 'nissan', 'mitsubishi', 'subaru', 'hyundai', 'toyota', 'mazda', 'bugatti', 'lamboghini', 'ferrari', ]
states=['tennessee', 'california', 'texas', 'florida', 'virginia', 'iowa', 'alaska', 'pennsylvania', ]
veggies=['broccoli', 'mushrooms', 'beets', 'potatoes', 'spinach', 'carrots', 'asparagus',]
planets=['earth', 'jupiter', 'mars', 'venus', 'saturn', 'neptune', 'mercury',]
word=''

while session=='y':
    CatChoice=input('What category do you want to guess from?\nFRUITS | VEGGIES | STATES | CARS | PLANTES \nIf you wish to finish your game press "q"\n').lower()  
    if CatChoice=="fruits":
        word=random.choice(fruits)
        game='y'
    elif CatChoice=="veggies":
        word=random.choice(veggies)
        game='y'
    elif CatChoice=="states":
        word=random.choice(states)
        game='y'
    elif CatChoice=="cars":
        word=random.choice(cars)
        game='y'
    elif CatChoice=="planets":
        word=random.choice(planets)
        game='y'
    elif CatChoice=="q":
        session='n'
        print("Thanks for playing!!")
    else:
        print("Yup... that's not a category")


 
    fails=7
    GuessWrong=[]
    GuessRight=[]
    guess=''

    while game == "y":

        print("GUESS THE CHARACTER")
        print("Your wrong answers...")
        print(GuessWrong)
        print("You have ", fails,  " attemps left")
        
        for letter in word:
            if letter in GuessRight:
                print(letter)
            else:
                print("_")
            
        guess = input("Guess a letter:\n")
        co()
        if guess  in word:
            GuessRight.append(guess)
            fails=fails-1
        elif guess not in word:
            GuessWrong.append(guess)
            fails=fails-1
            print("Wrong")

        if fails==0:
            print("You Loose")
            print("The word is: ", word)
            game="n"
                
        if word == GuessRight :
            print("You Win")
            print("The word is: ", word)
            game="n"
```
